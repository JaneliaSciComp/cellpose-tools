```python
import numpy as np
import zarr

def open_zarr(img_container_path, img_subpath, mode='r'):
    img_container = zarr.open(store=img_container_path, mode=mode)
    if img_subpath:
        return img_container[img_subpath]
    else:
        return img_container
```

```python
source_image_zarr_path = '/nearline/projtechres/GR_Fru_stitched_images/ZarrConvertForClaire/Gel1_Fru_594_dapi_1x_6x9c_VNC_stitched.zarr'
labels_zarr_path = '/Users/goinac/Work/HHMI/segmentation/testdata/Fru_test/test_data/cg-seg/labels.zarr'

source_image_dataset = 's0'
unmerged_labels_dataset = 'unmerged-labels/0'
merged_labels_dataset = 'merged-labels/0'

roi = (1260, 900, 800, 2010, 1620, 990)        # (xmin, ymin, zmin, xmax, ymax, zmax)
```

```python
source_image_zarr = open_zarr(source_image_zarr_path, source_image_dataset)
unmerged_labels_zarr = open_zarr(labels_zarr_path, unmerged_labels_dataset)
merged_labels_zarr = open_zarr(labels_zarr_path, merged_labels_dataset)
```

```python
xmin, ymin, zmin, xmax, ymax, zmax = roi
translate = (zmin, ymin, xmin)                                      # keeps napari's displayed coordinates matching the full volume, not the crop

source_nDapi_ch = source_image_zarr[1, zmin:zmax, ymin:ymax, xmin:xmax]
unmerged_labels = unmerged_labels_zarr[zmin:zmax, ymin:ymax, xmin:xmax]
merged_labels = merged_labels_zarr[zmin:zmax, ymin:ymax, xmin:xmax]
```

```python
unique_unmerged_labels = np.unique(unmerged_labels)
print("Unique unmerged labels", unique_unmerged_labels)

unique_merged_labels = np.unique(merged_labels)
print("Unique merged labels", unique_merged_labels)
```

```python
import napari

viewer = napari.Viewer()
viewer.add_image(source_nDapi_ch, name="DAPI", colormap="cyan", translate=translate)
viewer.add_labels(unmerged_labels, name="Unmerged masks", opacity=0.8, translate=translate)   # toggle on to compare against the merged result below
viewer.add_labels(merged_labels, name="Merged masks", opacity=0.8, translate=translate)

napari.run()
```
